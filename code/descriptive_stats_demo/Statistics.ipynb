```python
# Import all packages
import numpy as np
import pandas as pd
from pandas import Series

import matplotlib
import matplotlib.pyplot as plt

import scipy
import seaborn as sns
%matplotlib inline
```

```python
# Load data into notebook
# Use the pandas function '.read_excel' - name of excel sheet goes in ()
data = pd.read_excel("AminoAcid_Prop.xlsx")
data
```

```python
# Check the dimensions of your dataset to make sure all the data is there
# 20 rows/observations and 10 columns/variables
data.shape
```

```python
# Check to see all variable types are correct 
# int = +/- whole numbers (no decimals)
# float = real numbers (with decimals)
# object = boolean values - string
data.dtypes
```

```python
# Removing the index column --> the 'single_letter' will become the index values
# Good thing bc it is a unique identifier for the rows
data = pd.read_excel("AminoAcid_Prop.xlsx", index_col=0)
data
```

```python
# method '.columns' will show you the names of the columns
# Notice that single_letter is not included as a column/variable
data.columns
```

```python
# Use the function '.index' to check the index values. In this case, our index is the single letters
data.index
```

```python
# We can search for information on a specific row
# method '.loc' searches dataframe by label 
data.loc['A']
```

```python
# We can call out indexes using integers
# Method '.iloc' will allow you to call the index by a number instead of a letter
data.iloc[2]
```

```python
# Since the variable 'pka' is missing most of its values, we can remove that column
# Use '.drop' method since it is a df method 
# 'axis=1' tells pandas that we want to drop a column (pka). If we want to drop a row, we would use 'axis=0'
# 'inplace=True' means we want this operation to occur and we want it to replace the original dataset
data.drop(['pka'], axis=1, inplace=True)
```

```python
# Double check to make sure that the variable pka was removed
data
```

```python
# Double check dimensions
data.shape
```

```python
# The method '.describe' will provide summary statistics for columns containing numeric data
data.describe()
```

```python
# Check the median for each variable
data.median()
```

```python
data.mean()
```

```python
# Correlations - measures that show whether or how strongly one or more variables are related
# Default is a pearsons correlation
corr = data.corr()
corr
```
