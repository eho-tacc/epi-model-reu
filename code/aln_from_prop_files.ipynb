```python
import pandas as pd
import os
```

```python
sp_properties = {
    'dog': '../data/proc/resi_prop_delta_Canis_lupus_familiaris.csv',
    'chicken': '../data/proc/resi_prop_delta_Gallus_gallus.csv',
    'cow': '../data/proc/resi_prop_delta_Bos_taurus.csv',
    'human': '../data/proc/resi_prop_delta_Homo_sapiens.csv',
    'mouse': '../data/proc/resi_prop_delta_Mus_musculus.csv',
    'cat': '../data/proc/resi_prop_delta_Felis_catus.csv',
    'horse': '../data/proc/resi_prop_delta_Equus_caballus.csv'
}
```

```python
sp_dfs = []
for key, val in sp_properties.items():
    spd = pd.read_csv(val, comment='#')
    spd['species'] = key
    sp_subset = spd[['resi_compare', 'iso1_one_idx', 'species']]
    sp_dfs.append(sp_subset)
    
all_species = pd.concat(sp_dfs)
```

```python
all_species_complete = all_species.dropna()
```

```python
all_species_complete.head()
```

```python
aln_wide = all_species_complete.pivot(index='species', columns='iso1_one_idx', values='resi_compare')
```

```python
aln_wide.columns = [''.join(str(col)).strip() for col in aln_wide.columns.values]
```

```python
aln_wide = aln_wide.reset_index()
```
